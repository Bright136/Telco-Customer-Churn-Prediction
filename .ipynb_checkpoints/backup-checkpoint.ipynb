```python
def date_extracts(data):
    data['Year'] = data.index.year
    data['Month'] = data.index.month
    data['DayOfMonth'] = data.index.day
    data['DaysInMonth'] = data.index.days_in_month
    data['DayOfYear'] = data.index.day_of_year
    data['DayOfWeek'] = data.index.dayofweek
    data['Week'] = data.index.weekofyear
    data['Is_weekend'] = np.where(data['DayOfWeek'] > 4, 1, 0)
    data['Is_month_start'] = data.index.is_month_start.astype(int)
    data['Is_month_end'] = data.index.is_month_end.astype(int)
    data['Quarter'] = data.index.quarter
    data['Is_quarter_start'] = data.index.is_quarter_start.astype(int)
    data['Is_quarter_end'] = data.index.is_quarter_end.astype(int)
    data['Is_year_start'] = data.index.is_year_start.astype(int)
    data['Is_year_end'] = data.index.is_year_end.astype(int)
    data.reset_index(inplace=True)
    data.drop(columns=['description'], inplace=True)
```

```python
# create a function to fill null values in the oil price column
def impute_oil_missing_values(data):
    data['Oil_prices'] = data['Oil_prices'].interpolate(limit_direction ='backward').interpolate(method ='linear', limit_direction ='forward')
    total_missing_values = data['Oil_prices'].isnull().sum()
    print(total_missing_values)
```

```python
# create a function to define the type of day
def type_of_day(row):
    if row.Holiday_type is np.NaN:
        if row.Is_weekend == 0:
            return 'Workday'
        else:
            return 'No work'
    elif row.Holiday_type == 'Transfer' or row.Holiday_transferred == 'True':
        return 'Transferred holiday'
    elif row.Holiday_type == 'Additional' or row.Holiday_type == 'Bridge':
        return 'Additional Holiday'
    elif row.Holiday_type == 'Work Day':
        return 'Workday'
    else:
        return row.Holiday_type
```

```python
# create a typeofday column in train data 
train_data['TypeOfDay'] = train_data.apply(type_of_day, axis=1)
train_data.isnull().sum()
```

```python
# create a typeofday column in train data 
test_data['TypeOfDay'] = test_data.apply(type_of_day, axis=1)
test_data.isnull().sum()
```

```python
# create a function that assigns 0 or 1 to payday
def payday(row):
    if row.DayOfMonth == 15 or row.Is_month_end == 1:
        return 1
    else:
        return 0
```

## Features Scaling and Encoding

```python
# Use pandas.DataFrame.drop_duplicates method
y_train = train_set['Sales'].copy()
x_train = train_set.drop(columns=['Sales'])
numerical_train = x_train[['Oil_prices', 'Item_onpromo']]
```

```python
y_val = validation_set['Sales'].copy()
x_val = validation_set.drop(columns=['Sales'])
```

```python
num_cols = list(numerical_train.columns)
cat_cols = ['Family', 'Store_city', 'Store_type', 'Store_state', 'TypeOfDay', 'Holiday_level', 'Holiday_city']
```

```python
from sklearn.impute import SimpleImputer

# Separate the columns into categorical and numerical columns
numerical_attributes = list(num_cols)
categorical_attributes = cat_cols

# cat_imputer = SimpleImputer(strategy="most_frequent")
# le_encoder = LabelEncoder()

# create a numerical pipeline to encode and impute the missing values in the numerical columns
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy="mean")),('std_scaler', StandardScaler())])

# create a categorical pipeline to encode and impute the missing in the numerical columns
cat_pipeline = Pipeline([('imputer',SimpleImputer(strategy="most_frequent")),('cat_encoder', OrdinalEncoder())])
```

```python
# use the numerical pipeline to transform the dataframe 
x_train[numerical_attributes] = num_pipeline.fit_transform(x_train[numerical_attributes])
x_val[numerical_attributes] = num_pipeline.fit_transform(x_val[numerical_attributes])
test_data[numerical_attributes] = num_pipeline.fit_transform(test_data[numerical_attributes])
```

```python
# use the categorical pipeline to transform the dataframe 
x_train[categorical_attributes] = cat_pipeline.fit_transform(x_train[categorical_attributes])
x_val[categorical_attributes] = cat_pipeline.fit_transform(x_val[categorical_attributes])
test_data[categorical_attributes] = cat_pipeline.fit_transform(test_data[categorical_attributes])
```
